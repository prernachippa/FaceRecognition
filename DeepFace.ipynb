```python
from deepface import DeepFace
import pandas as pd

# Paths to test images and reference dataset
test_image_paths = ['Documents/dataset/test.jpg']
reference_dataset_path = "Documents/dataset/dataset"

# Initialize lists to store ground truth labels and predicted labels
true_labels = []
predicted_labels = []

# Define the function to extract the true label
def extract_true_label(image_path):
    # Implement the logic to extract the true label from the image path or external annotation
    # For now, let's return a placeholder value
    return "True_Label"  # Replace this with your actual implementation

# Loop over each test image
for test_image_path in test_image_paths:
    # Perform face recognition on the test image
    result = DeepFace.find(img_path=test_image_path, db_path=reference_dataset_path, enforce_detection=False, model_name='VGG-Face')

    # Extract predicted identity from the result
    if len(result) > 0:
        predicted_identity = result[0]['identity']
    else:
        predicted_identity = "Unknown"

    # Extract ground truth label
    true_label = extract_true_label(test_image_path)

    # Store ground truth label and predicted label
    true_labels.append(true_label)
    predicted_labels.append(predicted_identity)

# Convert lists to strings to ensure consistent data types
true_labels = [str(label) for label in true_labels]
predicted_labels = [str(label) for label in predicted_labels]

# Convert true_labels and predicted_labels to Pandas Series
true_labels_series = pd.Series(true_labels)
predicted_labels_series = pd.Series(predicted_labels)

# Calculate accuracy using Pandas vectorized operations
accuracy = (true_labels_series == predicted_labels_series).mean()
print("Accuracy:", accuracy)
```

```python
cap = cv2.VideoCapture (0)

while True:
    
    state, frame = cap.read()
    
    if state!=True:
        break

    res = DeepFace.find(frame, db_path='./Documents/dataset/dataset', enforce_detection=False, model_name='VGG-Face')
    
    if len(res[0]['identity']) > 15:
        name = res[0]['identity'][0].split('\\')[1]
        xmin = int(res[0]['source_x'][0])
        ymin = int(res[0]['source_y'][0])
        w = res[0]['source_w'][0]
        h = res[0]['source_h'][0]
        xmax = int(xmin + w)
        ymax = int(ymin + h)
        cv2.rectangle(frame, (xmin, ymin), (xmax, ymax), (0, 255, 0), 1)
        cv2.rectangle(frame, (xmin, ymin-25), (xmax, ymin), (255, 255, 255), -1)
        cv2.putText(frame, name, (xmin, ymin), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 0), 2, cv2.LINE_AA)
    else:
        for i in range(len(res[0]['identity'])):
            xmin = int(res[0]['source_x'][i])
            ymin = int(res[0]['source_y'][i])
            w = res[0]['source_w'][i]
            h = res[0]['source_h'][i]
            xmax = int(xmin + w)
            ymax = int(ymin + h)
            cv2.rectangle(frame, (xmin, ymin), (xmax, ymax), (0, 255, 0), 1)
            cv2.rectangle(frame, (xmin, ymin-25), (xmax, ymin), (255, 255, 255), -1)
            cv2.putText(frame, "Unknown", (xmin, ymin), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 0), 2, cv2.LINE_AA)

    cv2. imshow('attendance', frame)

    c = cv2.waitKey(1)
    
    if c == ord('q'):
        break
        
cap.release()
cv2.destroyAllWindows()
```
